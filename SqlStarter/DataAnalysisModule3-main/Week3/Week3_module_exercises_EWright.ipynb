```python
# Your query here
query7 = """ 
select
s.store_name 
,SUM(oi.quantity*oi.list_price) as tot_rev

from orders o
join stores s on o.store_id = s.store_id
join order_items oi on o.order_id = oi.order_id
join products p on p.product_id = oi.product_id
group by s.store_name 

"""
pd.read_sql(query7, connection)
```

```python
# Your query here
query8 = """ 
select
c.first_name
,c.last_name
,sum(oi.quantity*p.price) as total_spend

from customers c 
join orders o on o.customer_id = c.customer_id
join order_items oi on oi.order_id = o.order_id
join products p on oi.product_id = p.product_id
group by c.first_name,c.last_name
order by total_spend desc
limit 5;

"""
pd.read_sql(query8, connection)
```

```python
# Your query here
query9 = """ 

select
t.*
,row_number() over (partition by t.product_name order by t.quantity_sold) as rownum
from
(select
c.category_name
,p.product_name
,sum(oi.quantity) as quantity_sold
from products p 
join categories c on p.category_id = c.category_id
join order_items oi on oi.product_id = p.product_id
group by c.category_name, p.product_name
order by quantity_sold desc) as t
where rownum =1


"""
pd.read_sql(query9, connection)
```
